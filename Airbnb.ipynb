```python
import pandas as pd

# Load the Excel file with the correct sheet name
df = pd.read_excel(
    '/content/1730285881-Airbnb_Open_Data.xlsx',
    sheet_name='in',  # explicitly specify sheet
    engine='openpyxl'
)

# Clean column names and strip spaces
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Rename 'price' and 'service fee' columns
df.rename(columns={'price': 'price_$', 'service_fee': 'service_fee_$'}, inplace=True)

# Clean 'room_type' values
df['room_type'] = df['room_type'].astype(str).str.strip()

# Count each room type
property_types = df['room_type'].value_counts().to_frame(name='count')

# Display
print(property_types)
```

```python
room_type_bar=plt.bar(property_types.index,property_types.loc[:,"count"]);
plt.bar_label(room_type_bar, labels = property_types.loc[:,"count"], padding = 4 )
plt.ylim([0,50000]);
plt.xlabel('Room Type');
plt.ylabel('Room Type Count');
plt.title('Property Types and their count in the Dataset');
```

```python
hood_group = df["neighbourhood_group"].value_counts().to_frame()
hood_group
```

```python
hood_group_bar =plt.bar(hood_group.index, hood_group.loc[:,"count"])
plt.bar_label (hood_group_bar, labels = hood_group.loc[:,"count"], padding=4);
plt.ylim([0,40000]);
plt.xlabel('Neighborhood Groups');
plt.ylabel('Number of-Listings');
plt.xticks(rotation= 45);
plt.title('which Neighborbood Groups has the highest nuumber of listings');
```

```python
avg_price=df.groupby('neighbourhood_group')['price_$'].mean().sort_values(ascending = False).to_frame()
avg_price_bar= plt.bar(avg_price.index, avg_price.loc[:, 'price_$']);
plt.bar_label(avg_price_bar, labels=round(avg_price.loc[:,"price_$"], 2), label_type='edge', padding=4);

plt.ylim([0,700]);

plt.xlabel("Neighborhood Group");

plt.ylabel('Average Price per Listings ($)');

plt.xticks(rotation = 45);

plt.title("Average Price per Listings ($) in each freighborhood Group");
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
excel_file = '/content/1730285881-Airbnb_Open_Data.xlsx'

# Step 1 — Check sheet names
xls = pd.ExcelFile(excel_file)
print("Available sheets:", xls.sheet_names)

# Step 2 — Load correct sheet into a new variable
df_full = pd.read_excel(excel_file, sheet_name=xls.sheet_names[0], header=0)

# Step 3 — Check columns and construction years
print("Columns:", df_full.columns)
print("Unique construction years:", df_full['Construction year'].unique())

# Step 4 — Remove missing values
df_full.dropna(subset=['Construction year', 'price'], inplace=True)

# Step 5 — Group by construction year and plot average price
avg_price = df_full.groupby('Construction year')['price'].mean()
avg_price.plot(kind='line', marker='o')

plt.xlabel('Construction Year')
plt.ylabel('Average Price ($)')
plt.title('Average Price for properties per Construction Year')
plt.grid(True)
plt.show()

```

```python
import pandas as pd

df_fixed = pd.read_excel('/content/1730285881-Airbnb_Open_Data.xlsx', header=0)  # reload correct data

# Clean column names in df_fixed
df_fixed.columns = df_fixed.columns.str.strip().str.lower().str.replace(' ', '_')


print(df_fixed.shape)
print(df_fixed['host_name'].nunique())

# Now use df_fixed for plotting instead of df
hosts = df_fixed['host_name'].value_counts().head(10)

import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
bars = plt.bar(hosts.index, hosts.values, color='skyblue')
plt.bar_label(bars, labels=hosts.values, label_type='edge', padding=3)
plt.xlabel('Host Name')
plt.ylabel('Number of Listings')
plt.title('Top 10 Hosts by Listings Count')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
